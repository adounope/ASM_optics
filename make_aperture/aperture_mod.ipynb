```python
import numpy as np
import matplotlib.pyplot as plt
import src.math_tool as mt
import src.img_tool as it
import src.apertures as apertures
%load_ext autoreload
%autoreload 2
%matplotlib ipympl
```

```python
load_path = 'aperture/f32f45_angular8'
save_path = 'aperture/tmp/f32f45_angular8'
A_xy = np.load(f'{load_path}.npy')
plt.imshow(A_xy.T)
plt.show()
```

```python
A_xy_zoom = np.repeat(A_xy, 2, axis=0)
A_xy_zoom = np.repeat(A_xy_zoom, 2, axis=1)
plt.imshow(A_xy_zoom)
plt.show()
```

```python
# pad from 1024 to 4096, each side pad (4096-1024)/2=1536
target_size = 2048
A_xy_pad = it.img_pad(A_xy, (target_size-len(A_xy))//2, False)
plt.cla()
plt.imshow(A_xy_pad.T)
plt.show()
plt.savefig(f'{save_path}_pad_{target_size}.png', dpi=300, bbox_inches='tight')
np.save(f'{save_path}_pad_{target_size}.npy', A_xy_pad)
```

```python
# pad from 1024 to 4096, each side pad (4096-1024)/2=1536
target_size = 2048
A_xy_zoom_pad = it.img_pad(A_xy_zoom, (target_size-len(A_xy_zoom))//2, False)
plt.cla()
plt.imshow(A_xy_zoom_pad.T)
plt.show()
plt.savefig(f'{save_path}_zoom_pad_{target_size}.png', dpi=300, bbox_inches='tight')
np.save(f'{save_path}_zoom_pad_{target_size}.npy', A_xy_zoom_pad)
```
